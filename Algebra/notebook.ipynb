Pandas online tečajevi csv

```python
import pandas as pd

data = pd.read_csv('online_tecaj.csv')
data
```

```python
data.isnull().sum()
```

```python
data.nunique()
```

```python
data['course_title'].duplicated().any()
```

```python
data.drop_duplicates(subset=["course_title"], keep="last")
```

```python
data_paid = data[data['is_paid'] == True].copy()
data_free = data[data['is_paid'] == False].copy()

df_data_free_lvl = data_free.set_index('course_title').level
df_data_paid_lvl = data_paid.set_index('course_title').level
data_free
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
plt.Figure(figsize=(15, 10))
sns.countplot(y=df_data_free_lvl, order=df_data_free_lvl.value_counts().index[:3], palette="Spectral")
```

```python
plt.Figure(figsize=(15, 10))
sns.countplot(y=df_data_paid_lvl, order=df_data_paid_lvl.value_counts().index[:5], palette="Spectral")
```

```python
data.price.value_counts().sort_index(ascending=False).head(5).plot(kind='pie')
```
